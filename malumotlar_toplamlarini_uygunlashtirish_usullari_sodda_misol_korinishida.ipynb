```python
import pandas as pd
```

```python
import numpy as np
```

```python
A = pd.DataFrame({"Transport": ['Poyezd', 'Avtomobile', 'Kema'],
                  "Tezligi": [70, 180, 45]})
A
```

```python
B = pd.DataFrame({"Transport": ['Poyezd', 'Avtomobile', 'Tramvay'],
                  "Narxi": [98000, 10000, 65000]})
B
```

```python
merged = pd.merge(A, B)# inner join - ichki birlashish
```

```python
merged
```

```python
merged_outer = pd.merge(A, B, how='outer') # outer join - tashqi birlashish
merged_outer
```

```python
merged_left = pd.merge(A, B, how='left') # left join
merged_left
```

```python
merged_right = pd.merge(A, B, how='right')# right join
merged_right
```

```python
A = pd.DataFrame({"Transport_turi": ['Poyezd', 'Avtomobile', 'Kema'],
                  "Tezligi": [70, 180, 45]})
A
```

```python
B
```

```python
merged_2 = pd.merge(A, B, left_on='Transport_turi', right_on='Transport')
merged_2
```

```python
A = A.set_index("Transport_turi")
A
```

```python
B
```

```python
merged_3 = pd.merge(A, B, left_index=True, right_on='Transport')
merged_3
```

```python
B = B.set_index("Transport")
B
```

```python
merged_4 = pd.merge(A, B, left_index=True, right_index=True)
merged_4
```
