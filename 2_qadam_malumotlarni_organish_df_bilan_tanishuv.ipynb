```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://github.com/ageron/handson-ml2/blob/master/datasets/housing/housing.csv?raw=true"
df = pd.read_csv(url)
df.head()
```

```python
df.describe()
```

```python
%matplotlib inline
df.hist(bins=50, figsize=(20, 15))
plt.show()
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(df, test_size=0.2, random_state=35)
train_set.head(10)
```

```python
%matplotlib inline
df['median_income'].hist(bins=50, figsize=(9, 6))
plt.show()
```

```python
df['income_cat'] = pd.cut(df['median_income'], bins=[0, 1.5, 3, 4.5, 6, np.inf], labels=[1,2,3,4,5])

df['income_cat'].hist(figsize=(9, 6))
plt.show()
```

```python
df.sample(5)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
stratifiedshufflesplit = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in stratifiedshufflesplit.split(df, df['income_cat']):
  strat_train_set = df.loc[train_index]
  strat_test_set = df.loc[test_index]
strat_test_set
strat_train_set
```

```python
%matplotlib inline
strat_train_set['income_cat'].hist(figsize=(9, 6))
plt.show()
```

```python
strat_train_set.drop('income_cat', axis=1, inplace=True)

strat_test_set.drop('income_cat', axis=1, inplace=True)
```

```python
housing = strat_train_set.copy()
housing
```
