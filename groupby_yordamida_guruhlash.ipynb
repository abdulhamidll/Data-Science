```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/automobile_data.csv", index_col=0)
df.head()
```

```python
df.company.value_counts()
```

```python
df.groupby("company").mean()
```

```python
df.groupby("company").price.mean()
```

```python
df.groupby("company").agg([min, max, np.mean])
```

```python
df.groupby("company").price.agg([min, max, np.mean])
```
