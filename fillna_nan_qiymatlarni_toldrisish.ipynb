```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/automobile_data_nan.csv", index_col=0)
df.head()
```

```python
df.fillna(0)
```

```python
df.fillna({'horsepower': 90, 'wheel-base': 50})
```

```python
np.mean(df['wheel-base'])
```

```python
df.fillna({'horsepower': np.mean(df['horsepower']), 'wheel-base': np.mean(df['wheel-base'])})
```

```python
df.fillna(method='ffill')
```

```python
df.fillna(method='bfill')
```

```python
df[df.company=='toyota']
```

```python
df[df.company=='toyota'].fillna(method='bfill')
```

```python
df.fillna(df.mean())
```
