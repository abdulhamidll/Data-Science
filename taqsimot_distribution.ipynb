```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv",)
df = df[df.year > 2010]
df = df[df.transmission != 'Other']
df = df[df.fuelType != 'Other']
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=df,  x='price')
plt.show()
```

```python

sns.displot(data=df, x='price', hue="transmission", multiple='stack', height=6, aspect=2)
plt.show()
```

```python

sns.displot(data=df, x='price', col="transmission", height=6, aspect=2)
plt.show()
```
